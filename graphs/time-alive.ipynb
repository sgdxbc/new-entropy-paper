```python
%config InlineBackend.figure_formats = ['svg']
import pandas as pd
import seaborn as sns
from glob import glob
sns.set_theme(style="ticks", font_scale=1.2)

frames = [pd.read_csv(data) for data in glob('_data/time-alive_*.csv')]
data = pd.concat(frames, ignore_index=True)
data['Code'] = list(zip(data['fragment_n'], data['fragment_k']))
data
```

```python
# palette = sns.color_palette('flare', n_colors=5)
g = sns.relplot(data=data, 
    kind="line",
    x="duration", y="alive_count", hue="Code", 
    # palette='flare',
    height=2.56, aspect=2.) \
.set_xlabels("Time (second)") \
.set_ylabels("Persistent Fragment") \
.set(ylim=(0, 80))
g.axes[0][0].axhline(32, linestyle='--', color='gray')
g.savefig("time-alive.pdf")
```
